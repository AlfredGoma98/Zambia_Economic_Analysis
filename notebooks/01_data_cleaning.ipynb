# Zambia Economic Analysis - Data Cleaning
This notebook focuses on loading, inspecting, and cleaning macroeconomic data including GDP, exchange rates, and copper prices.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#Loading the dataset
df = pd.read_csv("C:/Users/user/Zambia_Economic_Analysis/data/raw/Zambia_Macro_Data.csv")
df.head()
```

```python
#understanding the data
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.copy()
```

```python
# Handling missing data
df['Inflation'] = df['Inflation'].interpolate(method='linear')
df['Debt_GDP'] = df['Debt_GDP'].interpolate(method='linear')
df['FDI_GDP'] =df['FDI_GDP'].interpolate()
```

```python
# Creating new features
# Population growth
df['Population_Growth'] = df['Population'].pct_change() * 100

# Lag features
df['GDP_Lag'] = df['GDP_Growth'].shift(1)
df['Inflation_Lag'] = df['Inflation'].shift(1)
```

```python
# Creating clean datasets
# Full dataset 
df_full = df.copy()
# ML dataset
df_ml = df[df['Year'] > 1990].copy()
df_ml = df_ml.dropna()
# debt_dataset
df_debt = df.dropna(subset=['Debt_GDP'])
```

```python
df_ml.isnull().sum()
```

```python
# Final validation
print("Full Data Missing:\n", df_full.isnull().sum())
print("\nML Data Missing:\n", df_ml.isnull().sum())
print("\nDebt Data Missing:\n", df_debt.isnull().sum())
```
