# Data Understanding

## 1. Dataset Overview
- Source: World Bank WDI + Copper Prices
- Time range: 1964–2024
- Rows: 61
- Columns: 7

## 2. Variable Dictionary
| Column | Meaning | Notes |
| --- | --- | --- |
| **Year** | Calendar year | 1964–2024 |
| **GDP_Growth** | Annual GDP growth (%) | Continuous, safe for forecasting |
| **Inflation** | CPI inflation (%) | Only available from 1986 onward |
| **Debt_GDP** | Public debt as % of GDP | Reliable from 1990 onward |
| **FDI_GDP** | Foreign direct investment (% of GDP) | Missing early years |
| **Population** | Total population | Complete |
| **Exchange_Rate** | Kwacha per USD | Very small values pre‑1990 due to rebasing |
| **Copper_Price** | Copper price in USD per pound | Complete |
